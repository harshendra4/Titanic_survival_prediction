```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df
```

```python
df2 = df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
```

```python
df2
```

```python
df_tgt=df2.Survived
df_tgt
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex=LabelEncoder()
```

```python
df2['Sex_n']=le_Sex.fit_transform(df['Sex'])
df2
```

```python
df2=df2.drop(['Sex','Survived'], axis='columns')
df2
```

```python
df2.isnull().values.any()
```

```python
df2['Age'].isnull().values.any()
```

```python
df2['Age'].fillna(value=df2['Age'].mean(), inplace=True)
df2
```

```python
df2['Age'].isnull().values.any()
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(df2,df_tgt)
```

```python
model.score(df2, df_tgt)
```
